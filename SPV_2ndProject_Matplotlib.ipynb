```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
```

```python
#loading and importing the data

df = pd.read_csv('football-players-drop-duplicates.csv')
df
```

```python
type(df)
```

```python
df.dtypes
```

```python
# Line grapph
df['Market Place Value in Million'].plot()

plt.title('Info of data')
plt.xlabel('X axis')
plt.ylabel('VAlue in Million')

plt.savefig('Figures/Info of Value.png')
```

```python
x = df['Skill Moves']
y = df['Total Performance']

fig,ax = plt.subplots(figsize=(5,4))
ax.hist(x, color = 'c')


plt.title('Total Performance Over skill moves')
plt.xlabel('Skill Moves')
plt.ylabel('Total Performance')
plt.show()

plt.savefig('Figures/Total Performance Over skill moves.png')
```

```python
df.plot(x = 'Age', 
        y = 'Skill Moves', 
        kind = 'bar', 
        color = 'pink',
        title = 'Skill Moves over Age vartically')
plt.savefig('Figures/Skill Moves over Age vartically.png')
```

```python
df.plot(x = 'Age', 
        y = 'Skill Moves', 
        kind = 'barh', 
        color = 'm',
        title = 'Skill Moves over Age Horizontally')
plt.savefig('Figures/Skill Moves over Age Horizontally.png')
```

```python
# scatter plot
df.plot(x = 'Total Performance', 
        y = 'Market Place Value in Million', 
        kind = 'scatter', 
        color = 'g',
        title = 'Performance based Makret place value for players')
plt.savefig('Figures/Performance based Makret place value for players.png')
```

```python
plt.style.use('ggplot')
x = df['Total Performance']
y = df['Market Place Value in Million']

fig,ax = plt.subplots(figsize=(5,4))
ax.hist(x)


plt.title('Price value for Performance')
plt.xlabel('Total Performance')
plt.ylabel('Market Place Value in Million')
plt.show()

plt.savefig('Figures/Price value for Performance.png')
```

```python
plt.style.use('classic')
df.plot(x = 'Age', 
        y = 'Total Performance', 
        kind = 'bar', 
        color = 'pink',
        title = 'Total Performance by Age')
plt.savefig('Figures/Total Performance by Age.png')
```
