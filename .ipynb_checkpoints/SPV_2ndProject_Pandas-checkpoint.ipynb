```python
#Manipulating Data (preprocessing data)

df['Name'] = df['Name'].str.lower()
df
```

```python
df_missing = pd.read_csv('football-players-missing data.csv')
df_missing
```

```python
#filling the missing data , implace atrribute

df_missing['Overall'].fillna(df_missing['Overall'].mean(),inplace=True)
df_missing
```

```python
df_missing['Height (cm)'].fillna(df_missing['Height (cm)'].mean(),inplace=True)
df_missing
```

```python
#removing missing value
df_missing.dropna(inplace=True)
df_missing
```

```python
#converting the data frame into csv file
df_missing.to_csv('football-players-missing-dropped.csv')
```

```python
#Creating a new column using the existing data
new_column = pd.Series([4,5,5,1,4,4,4,3,3,1,2,1,2,3,3])
df['Skill Moves'] = new_column
df
```

```python
df["Skill Moves"].fillna(3,inplace=True)
df
```

```python
#creating a column using python list

penalties = [75,85,81,40,79,86,82,85,75,11,88,40,88,75,85,86,90,79,25,27,40,85,47,83]
df["Total penalties of their career"] = penalties
df
```

```python
# Height is converting cm to inch
df["Height (cm)"]=df["Height (cm)"]/2.54
df
```

```python
# renaming columns
df = df.rename(columns={'Height (cm)' : 'Height (inch)'})
df
```

```python
df = df.rename(columns={'Overall' : 'Total Performance'})
df
```

```python
#Removing a column

df.drop("Skill Moves",axis=1)
```

```python
#Removing a row

df.drop(index = 5,axis=0)
```

```python
# Replacing the expressions
df['Market Place Value in Million'] = df['Market Place Value in Million'].str.replace('€','')
df
```

```python
# show the type of each column

df.dtypes
```

```python
# Converting the Market Place Value to integer

df['Market Place Value in Million'] = df['Market Place Value in Million'].str.replace('[\€\,]|\.\d*','').astype(int)
  
```

```python
# now see the change
df.dtypes
```
